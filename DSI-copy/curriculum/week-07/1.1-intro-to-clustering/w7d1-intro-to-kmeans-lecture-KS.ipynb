### My First<sup>TM</sup> Clustering Algorithm

1. Pick a value for k (the number of clusters to create)
2. Initialize k 'centroids' (starting points) in your data
3. Create your clusters. Assign each point to the nearest centroid. 
4. Make your clusters better. Move each centroid to the center of its cluster. 
5. Repeat steps 3-4 until your centroids converge.

# Metrics for assessing your clusters

**Inertia** -- sum of squared errors for each cluster
- ranges from 0 to very high values
- low inertia = dense clusters

**Silhouette Score** -- measure of how far apart clusters are
- ranges from -1 to 1
- high silhouette Score = clusters are well separated


**Inertia** -- sum of squared errors for each cluster
- low inertia = dense cluster

$$\sum_{j=0}^{n} (x_j - \mu_i)^2$$

where $\mu_i$ is a cluster centroid. (K-means explicitly tries to minimize this.)

`.inertia_` is an attribute of sklearn's kmeans models.


**Silhouette Score** -- measure of how far apart clusters are
- ranges from -1 to 1
- high silhouette Score = clusters are well separated

The [definition is a little involved](https://en.wikipedia.org/wiki/Silhouette_(clustering)), but intuitively the score is based on how much closer data points are to their own clusters than to the nearest neighbor cluster.

We can calculate it in sklearn with `metrics.silhouette_score(X_scaled, labels, metric='euclidean')`.

```python
%matplotlib inline 

import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from sklearn import cluster, datasets, preprocessing, metrics
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

# Group Practice
Let's do some clustering with the iris dataset.

```python
# Check out the dataset and our target values
df = pd.read_csv("assets/datasets/iris.csv")
print df.Name.value_counts()
df.head(5)
```

Our dataset has the categorical column Name in it, so we'll to convert this columns to numeric data for the k-means algorithm to accept it. 

Last week we used sklearn's `LabelEncoder` function to create numeric target values for the car dataset

For maximum clarity, we'll use a simple if-elif-else statement here. The Name column has Iris-setosa, Iris-virginica, and Iris-versicolor as attributes.

```python
def name_to_numeric(x):
    if x=='Iris-setosa':
        return 1
    elif x=='Iris-virginica':
        return 2
    else:
        # Note that this is risky in real life: 
        # how would we know if we had NaNs or other invalid attributes in the Name column?
        return 3
```

```python
df['name_num'] = df['Name'].apply(name_to_numeric)
del df['Name']
```

```python
df.head(2)
```

Now let's plot the data to see the distributions:

```python
cols = df.columns[:-1]
plt.style.use('fivethirtyeight')
sns.set_style('whitegrid')
sns.pairplot(df[cols])
```

Next, since each of our features have different units and ranges, let's do some preprocessing:

```python
X_scaled = preprocessing.MinMaxScaler().fit_transform(df[cols])
```

```python
pd.DataFrame(X_scaled, columns=cols).describe()
```

Now that we've formatted our data and understand it's structures, we can finally go ahead and cluster.

We're going to set k at two given behavior we were seeing above in our graphs.

```python
k = 2
kmeans = cluster.KMeans(n_clusters=k)
kmeans.fit(X_scaled)
```

We can use Scikit's built-in functions to determine the locations of the labels, centroids, and cluster inertia:

```python
# labels = cluster labels (so should have 2 since we specified k=2)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
inertia = kmeans.inertia_
```

And to compute the clusters' silhouette coefficient:


```python
metrics.silhouette_score(X_scaled, labels, metric='euclidean')
```

```python
inertia
```

...and we're done! You've completed your first clustering analysis.

Let's see how it looks. First, let's put the labels columns into our dataframe

```python
df['label'] = labels
df.head()
```

```python
cols = df.columns[:-1]
sns.pairplot(df, x_vars=cols, y_vars= cols, hue='label')
```

Just for comparison, here's the data colored by name of the plant.
