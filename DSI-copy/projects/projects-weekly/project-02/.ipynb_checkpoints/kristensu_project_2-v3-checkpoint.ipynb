```python
#1: Replace all * with np.nan
def no_asterisk(x):
    if x == '*':
        return np.nan
    else:
        return x

df.applymap(no_asterisk)
```

```python
#2: Reformat 'time' and convert from dtype object to timedelta in seconds
df['time'] = df['time'].str.replace(',',':')
df['time'] = df['time'].str.strip('AM')
df['time'] = pd.to_timedelta(df['time'], unit = 's')
```

```python
df['time'].head()
```

```python
#3: Convert dates ('date.entered', 'date.peaked') from dtype object to datetime
df['date.entered'] = pd.to_datetime(df['date.entered'], infer_datetime_format=True)
df['date.peaked'] = pd.to_datetime(df['date.peaked'], infer_datetime_format=True)
```

```python
print df['date.entered'].head(3)
print df['date.peaked'].head(3)
```

```python
#4: Rename 'artist.inverted' column to 'artist' 
df.rename(columns={'artist.inverted':'artist'}, inplace=True)
df['artist'].head(3)
```

```python
#5: Rename all column names with '.' to '_'
df.columns = [x.replace('.', '_') for x in df.columns]
df.columns.values
```

```python
#6: Convert week columns from dtype object to dtype numeric
for i in range(7,len(df.columns)):
    df[df.columns[i]] = pd.to_numeric(df[df.columns[i]],errors='coerce')
```

```python
df.dtypes
```

```python
#7a: Determine what genre names are
print df['genre'].nunique()
print df['genre'].unique()
print ""
print df['genre'].value_counts()
```

```python
#7b: Consolidate genre names
df['genre'].replace("Rock'n'roll", 'Rock', inplace=True)
df['genre'].replace('R & B', 'R&B', inplace=True)
print df['genre'].value_counts()
```

```python
#8: Examine how many artists there are and how often each appears in df
#To determine if there are artists with multiple hits
df['artist'].value_counts()
```

```python
df.head()
```

```python
#8: Reshape the data so that weeks become consolidated into one column (instead of each week spread as its own column)/
# and each week's rank becomes its own column
ndf = pd.melt(df, id_vars=['year', 'artist', 'track', 'time', 'genre', 'date_entered', 'date_peaked'], \
        var_name='week', value_name='list_rank')
```

```python
ndf.head()
```

```python
#9: Reformat 'week' from dtype object to int
ndf['week'] = ndf['week'].str.extract('x(\d+)').astype(int)
ndf['week'].head()
```

```python
#10: Sort the df by artist, track and then week
ndf.sort_values(by=['artist','track','week'], inplace=True)
ndf.head()
```

```python
#Double check to see sort worked properly
ndf.iloc[145:155,:]
```

```python
#11: Drop all the rows in the df that have NaN values
ndf.dropna(inplace=True)
ndf.iloc[145:155,:]
```

```python
#Dropping NaN values reduced index length from 24092 to 5307
ndf.index
```

```python
#12: Create a column that gives the date of each week listed in 'week' column 
#This also makes date formats comparable for plotting
ndf['week_date'] = ndf['date_entered'] + pd.TimedeltaIndex(ndf['week'] - 1, unit='W')
```

```python
ndf.head(3)
```

```python
#13: Create a column that gives the time to took for a track from entering the charts to peaking
ndf['to_peak'] = (ndf['date_peaked'] - ndf['date_entered']).apply(lambda x: x).astype(np.timedelta64(1,'W'))
```

```python
#pivot for no. of weeks on chart, for tracks, genres, time

max_week_tracks = pd.pivot_table(ndf, index=["track"], values=['week'], aggfunc=np.max) 
max_week_tracks.sort_values(by = ['week'], inplace=True)

max_week_genre = pd.pivot_table(ndf, index=["genre"], values=['week'], aggfunc=np.max) 
max_week_genre.sort_values(by = ['week'], inplace=True)

max_week_time = pd.pivot_table(ndf, index=["time"], values=['week'], aggfunc=np.max) 
max_week_time.sort_values(by = ['week'], inplace=True)
```

```python
max_week_genre.head()
```

```python
# pivot for change in rank, for genres/track and rank

delta_peak_genre_tracks = pd.pivot_table(ndf, index=['genre','track'], values=['list_rank'], \
                                                                   aggfunc=[np.amin, np.amax] ) 
delta_peak_genre_tracks.reset_index()
delta_peak_genre_tracks['difference'] = delta_peak_genre_tracks['amax'] - \
                                                                delta_peak_genre_tracks['amin'] 

delta_peak_genre_tracks.head()
```

```python
# pivot for change in rank and time

delta_peak_time = pd.pivot_table(ndf, index=["time"], values=['list_rank'], \
                                                                   aggfunc=[np.amin, np.amax] ) 
delta_peak_time.reset_index()

delta_peak_time['difference'] = delta_peak_time['amax'] - delta_peak_time['amin']

delta_peak_time.head()
```

```python
#pivot for average no. of weeks on chart, for time

#avg_max_week_genre.sort_values(by = ['week'], inplace=True)
avg_max_week_time = pd.pivot_table(ndf, index=["time"], values=['week']) 
avg_max_week_time.sort_values(by = ['week'], inplace=True)

avg_max_week_time.head()
```

```python
# pivot for avg. number of weeks rise to peak by tracks

avg_week_to_peak_tracks = pd.pivot_table(ndf, index=["track"], values=['to_peak']) 
avg_week_to_peak_tracks.sort_values(by = ['to_peak'], inplace=True)
avg_week_to_peak_tracks.reset_index()

avg_week_to_peak_tracks.tail()
```

```python
# pivot for avg. number of weeks rise to peak by genres
avg_week_to_peak_genre = pd.pivot_table(ndf, index=["genre"], values=['to_peak']) 
avg_week_to_peak_genre.sort_values(by = ['to_peak'], inplace=True)
avg_week_to_peak_genre.reset_index()
```

```python
# pivot for avg. number of weeks rise to peak by time
avg_week_to_peak_time = pd.pivot_table(ndf, index=["time"], values=['to_peak']) 
avg_week_to_peak_time.sort_values(by = ['to_peak'], inplace=True)
avg_week_to_peak_time.reset_index()

avg_week_to_peak_time.tail()
```

```python
def my_pivot(df, index, values, aggfunc, plt=False):
    piv = pd.pivot_table(df, index=index, values=values, aggfunc=aggfunc)
    #piv.sort_values(by = ['week'], inplace=True)
    #print piv
    if plt: piv.plot(title= 'Number of Weeks on BB Top 100 Chart',kind='hist', figsize=(8,4),\
                         bins=13, xticks=[5,10,15,20,25,30,35,40,45,50,55,60,65], legend=False)

my_pivot(ndf, index=["track"], values=['week'], aggfunc=np.max, plt=True)
```

```python
def my_pivot_rise(df, index, values, aggfunc, plt=False):
    piv = pd.pivot_table(df, index=index, values=values, aggfunc=aggfunc)
    #piv.sort_values(by = ['week'], inplace=True)
    #print piv
    if plt: piv.plot(title= 'Time (in weeks) to reach peak position', kind='hist', \
        figsize=(8,4), bins=13, xticks=[5,10,15,20,25,30,35,40,45,50,55,60,65], legend=False)


my_pivot_rise(ndf, index=["track"], values=['to_peak'], aggfunc=np.max, plt=True)
```
