```python
#Clean data step 1: Replace all the '*' values with NaN
def clean_data_1(x):
    if x == '*':
        return np.nan
    else:
        return x
```

```python
#Clean data step 2: Convert time into seconds
df_dict = {}
df_dict.update({x: df[x].dtypes for x in df.columns})
```

```python
#Clean data step 2: Convert time into seconds
df['time'] = df['time'].str.replace(',',':')
df['time'] = df['time'].str.strip('AM')
```

```python
df['time'].head()
```

```python
df['time'] = pd.to_timedelta(df['time'], unit = 's')
```

```python
#Clean data step #3: Convert dates ('date.entered', 'date.peaked') from dtype object to datetime
df.columns = [x.replace('.', '_') for x in df.columns]
```

```python
print df['date_entered'].head()
print df['date_peaked'].head()
```

```python
df['date_entered'] = pd.to_datetime(df['date_entered'], infer_datetime_format=True)
df['date_peaked'] = pd.to_datetime(df['date_peaked'], infer_datetime_format=True)
```

```python
#Clean data step #4: Convert week columns to numeric data type
for i in range(7,len(df.columns)):
    df[df.columns[i]] = pd.to_numeric(df[df.columns[i]],errors='coerce')

df.dtypes
```
