```python
indeed['Company'].apply(lambda x: str(x))
indeed['Company'].apply(lambda x: type(x))
```

```python
index = 0
for comp in indeed['Company']:
    indeed['Company'].iloc[index] = comp.replace("\n", "")
    index += 1
indeed
```

```python
#filter out sponsored

indeed = indeed.loc[indeed['Sponsored'] != True]
indeed = indeed.reset_index()
indeed
```

```python
indeed.drop(['index'], axis=1, inplace=True)
indeed
```

```python
#convert data types as needed

indeed['Job'].apply(lambda x: str(x))
indeed['Job'].apply(lambda x: type(x))
```

```python
indeed['Location'].apply(lambda x: str(x))
indeed['Location'].apply(lambda x: type(x))
```

```python
indeed['Salary'].apply(lambda x: str(x))
indeed['Salary'].apply(lambda x: type(x))
```

```python
#obtain salary values from current salary strings (step 1 of 2 for this task)
index = 0
for sal_string in indeed['Salary']:
    if type(sal_string) == float:
        pass
    else:
        m = re.search('\[\<nobr\>(.+?)\<\/nobr\>\]', sal_string)
        if m:
            sal_string = m.group(1)
            indeed['Salary'].iloc[index] = sal_string.replace("[<nobr>", "")
    index += 1
            
indeed
```

```python
#obtain salary values from current salary strings (step 2 of 2 for this task)

index = 0
for sal_string in indeed['Salary']:
    if type(sal_string) == float:
        pass
    else:
        num_list = re.findall("\d{2,3},\d{3}", sal_string)
        int_list = map(lambda x: int(x.replace(",", "")), num_list)
        if len(int_list) == 0:
            indeed['Salary'].iloc[index] = np.nan
        else:
            one_sal = sum(int_list)/len(int_list)
            indeed['Salary'].iloc[index] = one_sal            
    index += 1
    
indeed
    
```

```python
#making dummy variable for high/low salary (using salary median as defining value)

salary_median = indeed['Salary'].median()
print salary_median
indeed["Sal_Stat"] = ""
index = 0

for sal in indeed['Salary']:
    if np.isnan(sal):
        indeed['Sal_Stat'].iloc[index] = np.nan
    elif sal >= salary_median:
        indeed['Sal_Stat'].iloc[index] = 1
    else:
        indeed['Sal_Stat'].iloc[index] = 0
    index += 1
        
indeed
```

```python
#based on output, I will have to eventually rebucket cities into regions in a state 
#(scraping scraped more cities than I intended, which will result in too many dummy variables)

indeed['Location'].value_counts()
```

```python
#removing zipcode and neighborhood information from location values/strings

index = 0
for loc in indeed['Location']:
    loc_list = re.findall("[a-zA-Z ]+, [A-Z]{2}", loc)
    for ind_loc in loc_list:
        indeed['Location'].iloc[index] = ind_loc
    index += 1
    
indeed
```

```python
indeed['Job'].value_counts()
```

```python
# Data Cleaning issues that will need to be addressed eventually: 
# if job title has two key words (e.g., senior and scientist) only first instance kept. As of now, any job title 
# with keywords lead, sr, senior will be bucketed as sr_type regardless of analyst and scientist keyword. Will probably
# need to make two seperate dummy variables: one for senior/manager titles, and one for actual job title (analyst, engineer, etc.)
# will have to address this later....

indeed['job_type'] = ""
index = 0

for job_string in indeed['Job']:
    if "Senior" in job_string:
        indeed['job_type'].iloc[index] = 'sr_type'
    elif "Sr" in job_string:
        indeed['job_type'].iloc[index] = 'sr_type'
    elif "Lead" in job_string:
        indeed['job_type'].iloc[index] = 'sr_type'
    elif "Analyst" in job_string:
        indeed['job_type'].iloc[index] = 'an_type'
    elif "Scientist" in job_string:
        indeed['job_type'].iloc[index] = 'sct_type'
    else:
        indeed['job_type'].iloc[index] = 'other'
    index += 1
        
indeed
```

```python
# drop rows with NaN salary values 

indeed = indeed.dropna()
indeed.head()
```
