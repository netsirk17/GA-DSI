## BEGIN MODELING...

```python
#Setting my X variables/model features....One set has features with large # of multiclass (X_1) and the other excludes
#those features (X_2)
X_1 = imdb.drop(['awards', 'imdbRating','title','tomatoMeter','lang_class', 'rating_class_i'], axis =1)
X_2 = imdb.drop(['awards', 'imdbRating','title','tomatoMeter','language', 'Production','director','genre', 'rating_class_i'], axis =1)
```

```python
#Y vars were set above, during EDA....
```

```python
X_1.fillna('Missing', inplace=True)
X_2.fillna('Missing', inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()
X_1['Production'] = le.fit_transform(X_1['Production'])
X_1['director'] = le.fit_transform(X_1['director'])
X_1['genre'] = le.fit_transform(X_1['genre'])
X_1['rating'] = le.fit_transform(X_1['rating'])
X_1['language'] = le.fit_transform(X_1['language'])

X_2['rating'] = le.fit_transform(X_2['rating'])
```

ANALYSIS PLAN:

1) Cross-Validated, DT and Bagging Regressor with y1 as continuous variable (r^2)

2) Cross-Validated, DT and Bagging Regressor with y1 as categorical variable (accuracy)

3) Compare score metrics and choose better model

4) Cross-Validated Bagging model (Classifier) with some bucketed features (accuracy)

5) Compare score metrics and choose better model

6) Then, determine feature importance using random forest (Feature Ranks using their importance scores)

7) GridsearchCV Bagging model with only important features (Best parameters, best mean accuracy score)

8) Test model on new data: Most popular movies on IMDb (ratings range from 5.3 - 9.5 and include None)
    (Confusion matrix, classification report, ROC curve/AUC)

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import BaggingClassifier, BaggingRegressor
from sklearn.cross_validation import cross_val_score

def do_cross_val_class(model):
    scores = cross_val_score(model, X_1, y1_cat, cv=5, n_jobs=-1, scoring='accuracy')
    return scores.mean(), scores.std()

def do_cross_val_reg(model):
    scores = cross_val_score(model, X_1, y1_cont, cv=5, n_jobs=-1, scoring='mean_squared_error')
    return scores.mean(), scores.std()
```

```python
# 1A) Cross-Validated Decision Tree with y1 as continuous variable

dtr = DecisionTreeRegressor()
do_cross_val_reg(dtr)
```

```python
# 2A) Cross-Validated Decision Tree with y1 as categorical variable

dt = DecisionTreeClassifier()
do_cross_val_class(dt)
```

```python
# 1B) Cross-Validated Bagging Regressor with y1 as continuous variable

bdtr = BaggingRegressor(DecisionTreeRegressor())
do_cross_val_reg(bdtr)
```

```python
# 2B) Cross-Validated Bagging Classifier with y1 as categorical variable

bdt = BaggingClassifier(DecisionTreeClassifier())
do_cross_val_class(bdt)
```

```python
# Compare score metrics and choose better model

"""
Of the last four models explored, CV Bagging Classifier had best cross val score (0.9188). 
"""
```

```python
# 5) Cross-Validated Bagging model (Classifier) with some bucketed features (accuracy)
        #Did not have time to recategorize variables with numerous classes (multi-class)
        #So instead, trying another model without those features (X_2 subset) and comparing performance

def do_cross_val_class_2(model):
    scores = cross_val_score(model, X_2, y1_cat, cv=5, n_jobs=-1, scoring='accuracy')
    return scores.mean(), scores.std()
```

```python
# 6) Compare score metrics and choose better model

print "CV Bagging Classifier with X_2: " + str(do_cross_val_class_2(bdt))
```

```python
'''
Score is lower than other bagging classifier model, but equivalent to decision tree classifier. Will do step # 7 with both subsets of features (X_1 and X_2) to compare. 
'''
```

```python
# 7) Then, determine feature importance using random forest (Feature Ranks using their importance scores)

#7 with X_1:

from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier

forest = RandomForestClassifier()

forest.fit(X_1, y1_cat)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
feature_importances = pd.DataFrame(forest.feature_importances_,
                                   index = X_1.columns,
                                    columns=['importance']).sort_values('importance',
                                                                        ascending=False)
print feature_importances.head()

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X_1.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X_1.shape[1]), indices)
plt.xlim([-1, X_1.shape[1]])
plt.show()
```

```python
#7 with X_2:

from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier

forest = RandomForestClassifier()

forest.fit(X_2, y1_cat)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
feature_importances = pd.DataFrame(forest.feature_importances_,
                                   index = X_2.columns,
                                    columns=['importance']).sort_values('importance',
                                                                        ascending=False)
print feature_importances.head()

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X_2.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X_2.shape[1]), indices)
plt.xlim([-1, X_2.shape[1]])
plt.show()
```

```python
# 8) GridsearchCV Bagging model with only important features (Best parameters, best mean accuracy score)

# Important features: runtime and awards_num

from sklearn.grid_search import GridSearchCV


params = {"base_estimator__max_depth": [3,5,10,20],
          "base_estimator__max_features": [None, "auto"],
          "base_estimator__min_samples_leaf": [1, 3, 5, 7, 10],
          "base_estimator__min_samples_split": [2, 5, 7],
          'bootstrap_features': [False, True],
          'max_features': [0.5, 0.7, 1.0],
          'max_samples': [0.5, 0.7, 1.0],
          'n_estimators': [2, 5, 10, 20],
         }
    

gsbdt = GridSearchCV(bdt, params, n_jobs=-1, cv=3)    
```

```python
X_important = X_2.drop(['rating', 'year', 'lang_class'], axis=1)

gsbdt.fit(X_important, y1_cat)
```
