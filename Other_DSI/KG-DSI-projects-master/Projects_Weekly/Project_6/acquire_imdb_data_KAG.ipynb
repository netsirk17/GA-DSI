```python
import requests
import bs4
from bs4 import BeautifulSoup, BeautifulStoneSoup
import urllib2
from IPython.display import HTML
import re
import numpy as np
import pandas as pd
import json
import urllib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# low 100 and top 250

def get_low_100():
    response = requests.get('http://www.imdb.com/chart/bottom')
    html = response.text
    low_entries = re.findall("<a href.*?/title/(.*?)/", html)
    return list(set(low_entries))

def get_top_250():
    response = requests.get('http://www.imdb.com/chart/top')
    html = response.text
    top_entries = re.findall("<a href.*?/title/(.*?)/", html)
    return list(set(top_entries))
```

```python
low_imdb = get_low_100()
top_imdb = get_top_250()
```

```python
low_and_top = low_imdb + top_imdb
print len(low_and_top)
```

```python
def get_entry(entry):
    res = requests.get('http://www.omdbapi.com/?i='+entry)
    if res.status_code != 200:
        print entry, res.status_code
    else:
        print '.',
    try:
        j = json.loads(res.text)
    except ValueError:
        j = None
    return j


def get_gross(entry):
    response = requests.get('http://www.imdb.com/title/'+entry)
    html = response.text
    try:
        gross_list = re.findall("Gross:</h4>[ ]*\$([^ ]*)", html)
        gross = int(gross_list[0].replace(',', ''))
        print '.',
        return gross
    except Exception as ex:
        print
        print ex, entry, response.status_code
        return None
```

```python
entries_dict_list = [get_entry(e) for e in low_and_top]
```

```python
grosses = [(e, get_gross(e)) for e in low_and_top]
```

```python
df1 = pd.DataFrame(grosses, columns=['imdbID', 'Gross'])
df1.head()
```

```python
imdb_df = pd.DataFrame(entries_dict_list)
```

```python
imdb_df.tail()
```

```python
results = []

for title in imdb_df['Title']:
    title = title.encode('utf8')
    title = title.replace(" ", "+")
    URL = "http://www.omdbapi.com/?t={title}&plot=full&r=xml&tomatoes=true".format(**locals())
    i = requests.get(URL)
    i_soup = BeautifulSoup(i.content, "lxml-xml")
    results.append(i_soup)
```

```python
def extract_everything(list_of_soup_files):
    """Extract desired movie info using BS4. Future: will refactor using modulo (%) 
    to pass in a list of desired tag/attribute names if possible..."""
    list_dict_350 = []
    for soup_file in list_of_soup_files:
        dict_350 = {}
        
        titles=soup_file.findAll('movie',{"title":True})
        for title in titles:
            dict_350['title'] = title['title']
        
        ratings=soup_file.findAll('movie',{"rated":True})
        for rating in ratings:
            dict_350['rating'] = str(rating['rated'])
        
        release_dates=soup_file.findAll('movie',{"released":True})
        for release in release_dates:
            dict_350['released'] = str(release['released'])
        
        run_times=soup_file.findAll('movie',{"runtime":True})
        for run_time in run_times:
            dict_350['runtime'] = str(run_time['runtime'])
        
        genres=soup_file.findAll('movie',{"genre":True})
        for genre in genres:
            dict_350['genre'] = str(genre['genre'])
        
        directors=soup_file.findAll('movie',{"director":True})
        for director in directors:
            dict_350['director'] = director['director'].encode('utf8')
        
        writers=soup_file.findAll('movie',{"writer":True})
        for writer in writers:
            dict_350['writer'] = writer['writer'].encode('utf8')
        
        actors=soup_file.findAll('movie',{"actors":True})
        for actor in actors:
            dict_350['actors'] = actor['actors'].encode('utf8')
        
        plots=soup_file.findAll('movie',{"plot":True})
        for plot in plots:
            dict_350['plot'] = plot['plot'].encode('utf8')
        
        languages=soup_file.findAll('movie',{"language":True})
        for language in languages:
            dict_350['language'] = str(language['language'])
        
        countrys=soup_file.findAll('movie',{"country":True})
        for country in countrys:
            dict_350['country'] = str(country['country'])
        
        awards=soup_file.findAll('movie',{"awards":True})
        for award in awards:
            dict_350['awards'] = str(award['awards'])
        
        metascores=soup_file.findAll('movie',{"metascore":True})
        for metascore in metascores:
            dict_350['metascore'] = str(metascore['metascore'])
        
        imdbRatings=soup_file.findAll('movie',{"imdbRating":True})
        for imdbRating in imdbRatings:
            dict_350['imdbRating'] = str(imdbRating['imdbRating'])
        
        imdbVotes=soup_file.findAll('movie',{"imdbVotes":True})
        for imdbVote in imdbVotes:
            dict_350['imdbVotes'] = str(imdbVote['imdbVotes'])
        
        imdbIDs=soup_file.findAll('movie',{"imdbID":True})
        for imdbID in imdbIDs:
            dict_350['imdbID'] = str(imdbID['imdbID'])
        
        tomatoMeters=soup_file.findAll('movie',{"tomatoMeter":True})
        for tomatoMeter in tomatoMeters:
            dict_350['tomatoMeter'] = str(tomatoMeter['tomatoMeter'])
    
        tomatoImages=soup_file.findAll('movie',{"tomatoImage":True})
        for tomatoImage in tomatoImages:
            dict_350['tomatoImage'] = str(tomatoImage['tomatoImage'])
                
        tomatoRatings=soup_file.findAll('movie',{"tomatoRating":True})
        for tomatoRating in tomatoRatings:
            dict_350['tomatoRating'] = str(tomatoRating['tomatoRating'])
                
        tomatoReviews=soup_file.findAll('movie',{"tomatoReviews":True})
        for tomatoReview in tomatoReviews:
            dict_350['tomatoReviews'] = str(tomatoReview['tomatoReviews'])
                
        tomatoFreshs=soup_file.findAll('movie',{"tomatoFresh":True})
        for tomatoFresh in tomatoFreshs:
            dict_350['tomatoFresh'] = str(tomatoFresh['tomatoFresh'])
                
        tomatoRottens=soup_file.findAll('movie',{"tomatoRotten":True})
        for tomatoRotten in tomatoRottens:
            dict_350['tomatoRotten'] = str(tomatoRotten['tomatoRotten'])
                
        tomatoConsensuses=soup_file.findAll('movie',{"tomatoConsensus":True})
        for tomatoConsensus in tomatoConsensuses:
            dict_350['tomatoConsensus'] = tomatoConsensus['tomatoConsensus'].encode('utf8')
                
        tomatoUserMeters=soup_file.findAll('movie',{"tomatoUserMeter":True})
        for tomatoUserMeter in tomatoUserMeters:
            dict_350['tomatoUserMeter'] = tomatoUserMeter['tomatoUserMeter'].encode('utf8')
                
        tomatoUserRatings=soup_file.findAll('movie',{"tomatoUserRating":True})
        for tomatoUserRating in tomatoUserRatings:
            dict_350['tomatoUserRating'] = str(tomatoUserRating['tomatoUserRating'])
                
        tomatoUserReviews=soup_file.findAll('movie',{"tomatoUserReviews":True})
        for tomatoUserReview in tomatoUserReviews:
            dict_350['tomatoUserReviews'] = str(tomatoUserReview['tomatoUserReviews'])
                
        DVDs=soup_file.findAll('movie',{"DVD":True})
        for DVD in DVDs:
            dict_350['DVD'] = str(DVD['DVD'])
                
        BoxOffices=soup_file.findAll('movie',{"BoxOffice":True})
        for BoxOffice in BoxOffices:
            dict_350['BoxOffice'] = str(BoxOffice['BoxOffice'])
                
        Productions=soup_file.findAll('movie',{"Production":True})
        for Production in Productions:
            dict_350['Production'] = str(Production['Production'])
            
        years=soup_file.findAll('movie',{"year":True})
        for year in years:
            dict_350['year'] = str(year['year'])
        
        list_dict_350.append(dict_350)        
    
    return list_dict_350

all_350_raw = extract_everything(results)
```

```python
imdb_350_raw = pd.DataFrame.from_records(all_350_raw)
imdb_350_raw = pd.merge(imdb_350_raw, df1)
imdb_350_raw.tail()
```
