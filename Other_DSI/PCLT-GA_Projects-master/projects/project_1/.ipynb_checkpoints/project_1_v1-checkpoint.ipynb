##### 1. What does the data describe?

The data show the mean SAT verbal and math scores, and participation rate, by state for the year 2001.

##### 2. Does the data look complete? Are there any obvious issues with the observations?


The data look complete insofar as there are values for all three variables for all 50 states and DC.  The participation rates, however, are very low for some states.  Given this, the mean scores may not be a reasonably accurate representation of the overall competence of the students in those states.  It may be the case that in states with low participation, the students self-select and only those who are most prepared take the SAT.  Therefore, the mean SAT scores may suggest greater overall academic competence that is the case.

##### 4. Load the data into a list of lists

```python
# open the csv file and create empty list for data
import csv
sat_csv_path = 'assets/sat_scores.csv'
data=[]

#read each line from csv into the data list
with open(sat_csv_path,'r') as f:
    reader=csv.reader(f)
    for row in reader:
        data.append(row)
```

##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
def stdev(num_list):
    avg=np.mean(num_list)
    sd=(np.mean([(x-avg)**2 for x in num_list]))**0.5
    return sd

print "Standard deviation for Rate is " +str(stdev(sat_data_dict["Rate"]))
print "Standard deviation for mean Verbal score by state is " +str(stdev(sat_data_dict["Verbal"]))
print "Standard deviation for mean Math score by state is " +str(stdev(sat_data_dict["Math"]))


```

## Step 4: Visualize the data

##### 14. Using MatPlotLib and PyPlot, plot the distribution of the Rate using histograms.

```python
# create list of rates
rates=sat_data_dict["Rate"]

#assign bin range and size
bins1=range(0,100,10)

#plot histogram for rates
plt.hist(rates,bins=bins1,color="darkorange")
plt.title("Distribution of Participation Rates")
plt.xlabel("Rate")
plt.ylabel("Frequency")
plt.show()
```

##### 15. Plot the Math distribution

```python
# create list of mean Math scores
math=sat_data_dict["Math"]

# plot histogram of mean math scores
plt.hist(math,bins=7,color="darkorange")
plt.title("Distribution of mean Math Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

##### 16. Plot the Verbal distribution

```python
# create list of mean Verbal scores
verbal=sat_data_dict["Verbal"]

# plot histogram of mean Verbal scores
plt.hist(verbal,bins=7,color="darkorange")
plt.title("Distribution of mean Verbal Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

##### 20. Are there any interesting relationships to note?

Generally, math and verbal scores are positively associated: states in which the average math score was relatively low tended also to have an average verbal score which was relatively low.  
The participation rate appears to be negatively associated with both average math and average verbal score.  In those states with low participation rates, the scores tend to be high.  This may indicate that those states with low participation rates may be those in which only the most prepared students take the SATs.

##### Repeat the charts in Seaborn

```python
#import seaborn library
import seaborn as sns


```

```python
import pandas as pd
# reimport data as dataframe to use for seaborn
sat_scores = pd.read_csv('assets/sat_scores.csv')
# create long dataset for facetgrids
sat_scores_long = pd.melt(sat_scores,
                      id_vars=['State','Rate'],
                      value_vars=['Verbal','Math'],
                      var_name='Variable',
                      value_name='Value')
sat_scores_long.head()
```

```python
#plot histogram of rates
fig, ax = plt.subplots()
fig.set_size_inches(6, 4)
sns.distplot(rates, bins=range(0,100,10), kde=False, rug=True,ax=ax)
plt.title("Distribution of Rates",fontsize=16)
```

```python

g = sns.FacetGrid(sat_scores_long, col='Variable',sharex=False)
g = g.map(sns.distplot, 'Value',kde=False, bins=10)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(4, 4)
sns.boxplot(rates,ax=ax, orient="v")
plt.title("Rates",fontsize=16)
```

```python
g = sns.FacetGrid(sat_scores_long, col='Variable')
g = g.map(sns.boxplot, 'Value',orient="v")
```

```python
g = sns.FacetGrid(sat_scores_long, hue="Variable", palette="Set1", size=5, hue_kws={"marker": ["o", "v"]})
g.map(plt.scatter, "Rate", "Value", s=100, linewidth=.5, edgecolor="white")
g.add_legend();
```
