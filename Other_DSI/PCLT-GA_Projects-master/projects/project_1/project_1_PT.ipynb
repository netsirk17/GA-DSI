##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
#define function to calculate standard deviation
def stdev(num_list):
    """Calculates and returns standard deviation
    Param:
    num_list: list of numbers
    """
    avg=np.mean(num_list)
    sd=(np.mean([(x-avg)**2 for x in num_list]))**0.5
    return sd

print "Standard deviation for Rate is " +str(stdev(sat_data_dict["Rate"]))
print "Standard deviation for mean verbal score by state is " +str(stdev(sat_data_dict["Verbal"]))
print "Standard deviation for mean math score by state is " +str(stdev(sat_data_dict["Math"]))
```

##### 20. Are there any interesting relationships to note?

Generally, math and verbal scores are positively associated: states in which the mean math score was relatively low tended also to have a mean verbal score which was relatively low.  
The participation rate appears to be negatively associated with both average math and average verbal scores.  In those states with low participation rates, the scores tend to be high.  It may be the case that in states with low participation, the students self-select and only those who are most prepared take the SAT.  Therefore, the mean SAT scores may suggest greater overall academic competence than is the case.

##### Repeat the charts in Seaborn

```python
#import seaborn library
import seaborn as sns
```

```python
import pandas as pd
# reimport data as dataframe to use for seaborn
sat_scores = pd.read_csv('assets/sat_scores.csv')
# create long dataset for facetgrids.  since rate is on a very different scale to scores, leave
# rate as a variable
sat_scores_long = pd.melt(sat_scores,
                      id_vars=['State','Rate'],
                      value_vars=['Verbal','Math'],
                      var_name='Subject',
                      value_name='Score')
```

```python
#plot histogram of rates
fig, ax = plt.subplots()
fig.set_size_inches(6, 4)
sns.distplot(rates, bins=range(0,100,10), kde=False, rug=True,ax=ax)
plt.title("Distribution of Rates",fontsize=16)
```

```python
# plot verbal and math scores in facets
g = sns.FacetGrid(sat_scores_long, col='Subject',sharex=False)
g = g.map(sns.distplot, 'Score', kde=False, bins=10)
```

```python
# box plot for rates
fig, ax = plt.subplots()
fig.set_size_inches(4, 4)
sns.boxplot(rates,ax=ax, orient="v")
plt.title("Rates",fontsize=16)
```

```python
# box plots for verbal and math scores
g = sns.FacetGrid(sat_scores_long, col='Subject')
g = g.map(sns.boxplot, 'Score',orient="v")
```

```python
# plot rate vs scores
g = sns.FacetGrid(sat_scores_long, hue="Subject", palette="Set1", size=5, hue_kws={"marker": ["o", "v"]})
g.map(plt.scatter, "Rate", "Score", s=100, linewidth=.5, edgecolor="white")
g.add_legend();
```
