```python
from __future__ import division
import requests
import bs4
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
import collections
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
import statsmodels.api as sm

from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, auc,\
roc_auc_score, roc_curve
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# functions to extract listing elements

# extract location
def extract_location_from_result(result):
    try:
        return result.find("span", class_="location").text
    except:
        return np.nan

# extract job title
def extract_jobtitle_from_result(result):
    try:
        return result.find(class_="jobtitle").text.strip()
    except:
        return np.nan
    
# extract location
def extract_company_from_result(result):
    try:
        return result.find("span", class_="company").text.strip()
    except:
        return np.nan

#extract salary 2
def extract_salary_from_result2(result):
    for i in result.findAll("div", class_="sjcl"):
        try:
            for j in i.find("div"):
                return j.strip()
        except:
            return np.nan

#extract salary
def extract_salary_from_result(result):
    for i in result.findAll("td", class_="snip"):
        try:
            for j in i.find("nobr"):
                return j.strip()
        except:
            return extract_salary_from_result2(result)
        
# extract description
def extract_description_from_result(result):
    try:
        return result.find("span", class_="summary").text.strip()
    except:
        return np.nan
```

```python
# search indeed.com pages and create a list of results

url_template = "http://www.indeed.com/jobs?q=data+scientist+%2420%2C000&l={}&start={}"
max_results_per_city = 400

my_cities = ['New+York', 'Chicago', 'San+Francisco', 'Austin', 'Atlanta','Boston','Los+Angeles','Seattle',\
            'Denver','Philadelphia','San+Diego']
results = []
page_count = 0
results_pp = []
for city in set(my_cities):
    for start in range(0,max_results_per_city,10):
        URL = url_template.format(city,start)
        r = requests.get(URL)
        indeed_soup = BeautifulSoup(r.content, "lxml")
        results.append(indeed_soup)
        page_count+=1
        results_pp.append(len(indeed_soup.findAll("div", class_="result")))                          
```

```python
# extract listing information into lists
location_list = []
job_list = []
company_list = []
salary_list = []
description_list= []
for page in results:
        for result in page.findAll("div", class_="result"):
            location_list.append(extract_location_from_result(result))
            job_list.append(extract_jobtitle_from_result(result))
            company_list.append(extract_company_from_result(result))
            salary_list.append(extract_salary_from_result(result))
            description_list.append(extract_description_from_result(result))
```

```python
# sometimes the number of results per city is inconsitent.  create df to calculate actual number of results per city
city_pages = []
for city in set(my_cities):
    for i in range(int(max_results_per_city/10)):
        city_pages.append(city)
city_count = pd.DataFrame({'city':city_pages, 'result_count':results_pp})
city_count = city_count.groupby(["city"])["result_count"].agg(np.sum).to_frame().reset_index()
```

```python
# create list of cities (location isn't always consistent)
city_list =[]
for city in set(my_cities):
    for i in range(city_count[city_count["city"]==city]["result_count"]):
        city_list.append(city)
```

```python
#create df of results
df = pd.DataFrame(zip(city_list,location_list,company_list,job_list,salary_list,description_list))
df.columns = ["city","location","company","role","salary","description"]
print df.shape
df.head(2)
```

```python
# replace none salaries with NA
def replace_none(x):
    if x ==None:
        return np.nan
    else:
        return x

df = df.applymap(replace_none)
df.head()
```

```python
# some of the listings are repeated - remove repeats
df.drop_duplicates(inplace=True)
df.shape
```

```python
df["city"].value_counts()
```

```python
# subset the df into entries with a salary listing, by year
sub = df[~df["salary"].isnull()]
sub = sub[sub["salary"].str.contains("year")]
sub.reset_index(inplace=True,drop=True)
sub["salary"] = sub["salary"].apply(lambda x: x.replace("a year",""))
sub["salary"]= sub["salary"].apply(lambda x: (str(x)).split('-'))

# create field for average salary if salary is a range
sub["lower_salary"] = sub["salary"].apply(lambda x: x[0].replace("$","").replace(",","")).astype(int)
sub["upper_salary"] = sub["salary"].apply(lambda x: x[-1].replace("$","").replace(",","")).astype(int)
sub["ave_salary"] = (sub["lower_salary"] + sub["upper_salary"])/2
```

```python
# plot ROC curve
roc = roc_curve(Y_test, logreg.predict_proba(X_test).T[1])
def getROCcurve(roc, title='ROC Curve', xlabel ='% of False Positives', ylabel = '% of True Positives'):
    
    fpr, tpr, thresholds = roc
    
    fig,ax = plt.subplots()
    ax.plot(fpr, tpr, label='ROC Curve')

    fig.set_size_inches(6, 6, forward=True)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(loc='best')
    
getROCcurve(roc, title='Predicting Salary', ylabel='% of high salary predicted as high salary',\
            xlabel='% of low salary predicted as high salary')
```

```python
# first create functions to print out metrics and ROC
def print_metrics(y_pred,Y_test,X_test,model):
    
    #print confusion matrix
    conmat = np.array(confusion_matrix(Y_test, y_pred, labels=[1,0]))
    confusion = pd.DataFrame(conmat, index=['High Salary', 'Low Salary'],\
                         columns=['Predicted High Salary','Predicted Low Salary'])
    print(confusion),"\n"

    # print metrics
    print "Accuracy:", accuracy_score(Y_test,y_pred)
    print "Precision:",precision_score(Y_test,y_pred)
    print "Recall:",recall_score(Y_test,y_pred)
    print "False Positive Rate:", confusion.iloc[1,0]/(confusion.iloc[1,0]+confusion.iloc[1,1])
    print "Area Under ROC Curve:",roc_auc_score(Y_test,model.predict_proba(X_test).T[1])
    print "\n"

    #print coefs
    print pd.DataFrame({"variable:":X_test.columns.values,"coef":model.coef_[0]})
    
def print_roc_graph(y_pred,Y_test,X_test,model):
        # plot ROC
    roc = roc_curve(Y_test, model.predict_proba(X_test).T[1])
    getROCcurve(roc, title='Predicting Salary', ylabel='% of high salary predicted as high salary',\
    xlabel='% of low salary predicted as high salary')
```
